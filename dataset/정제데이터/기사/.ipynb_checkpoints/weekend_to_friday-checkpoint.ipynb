```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
import os
news_list = [file for file in os.listdir('.') if 'csv' in file]
news_list
```

```python
def weekend_to_friday(file_name):
    df = pd.read_csv(file_name, index_col=0)

    for i in range(len(df)):
        now = datetime.strptime(df.loc[i,'date'], '%Y.%m.%d')
        if now.weekday() == 5:
            before_one_day = now - timedelta(days=1)
            df.loc[i,'date'] = str(before_one_day)[:10].replace('-','.')
        elif now.weekday() == 6:
            before_two_day = now - timedelta(days=2)
            df.loc[i,'date'] = str(before_two_day)[:10].replace('-','.')

    df.to_csv(file_name)
```

```python
for file_name in news_list:
    if file_name == '네이버_기사_크롤링_2020.01.01_2021.12.06.csv': 
        continue
    print(file_name)
    print('시작')
    weekend_to_friday(file_name)
    print('완료')
```

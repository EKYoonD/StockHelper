```python
# 기사 데이터들 가져오기
news_path = r'../../정제데이터/기사'
news_list = [file for file in os.listdir(news_path) if 'csv' in file]
news_list
```

```python
# 종목코드 데이터프레임 가져오기
stock_code_file = r'../../../static/data/stockList.txt'
stock_code_df = pd.read_csv(stock_code_file)
stock_code_df
```

```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('../../dataset/정제데이터/temp/삼성카드_기사_크롤링_2020.01.01_2021.12.06.csv', index_col=0)
df
```

```python
for i in range(len(df)):
    now = datetime.strptime(df.loc[i,'date'], '%Y.%m.%d')
    if now.weekday() == 5:
        before_one_day = now - timedelta(days=1)
        df.loc[i,'date'] = str(before_one_day)[:10].replace('-','.')
    elif now.weekday() == 6:
        before_two_day = now - timedelta(days=2)
        df.loc[i,'date'] = str(before_two_day)[:10].replace('-','.')
        
df[df['date'] == '2020.08.21']
```

```python
for i in range(len(df)):
    now = datetime.strptime(df.loc[i,'date'], '%Y.%m.%d')
    if now.weekday() == 5 or now.weekday() == 6:
        print(now)
```

```python
twitter = konlpy.tag.Twitter()
nlp = twitter   # Customized Konlpy


name_pos = {
    "Noun" : "명사",
    "Verb" : "동사",
    'Adjective' : '형용사'
}



def sep_pos(article):
    # 기사의 어휘소 분석
    morphs = nlp.pos(article, norm=True, stem=True)
     
    pos_words = []
                     
    for pos in name_pos:  # "Noun", "Verb"  통계검색할 품사들
        
        for mal in morphs:
            # 통계집계할 단어품사만 and 단어길이가 2글자 이상만
            if pos == mal[1]:  
                pos_words.append(mal)  # 단어 추가
           
    return pos_words




# malist = sep_pos(" ".join(df_chosun['정제본문']))

```

```python
okt = Okt()

mallist = okt.pos(article)
mallist_article = list()
for word in mallist : 
    if (word[1] in name_pos) & (len(word[0])>=2):
        mallist_article.append(word)
```

```python
word_article = [word[0] for word in mallist_article]
```

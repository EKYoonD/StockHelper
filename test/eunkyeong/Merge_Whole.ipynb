```python
import pandas as pd
import numpy as np
import datetime
```

```python
contents = pd.read_csv(r'C:\StockHelper_Project\StockHelper\dataset\삼성증권_감성점수.csv')
contents['date'] = pd.to_datetime(contents['date'], infer_datetime_format=True) 
stock_date_score = contents.groupby('date').mean()['Senti_Score']
```

```python
stock_date_score = pd.DataFrame(stock_date_score)
```

```python
import FinanceDataReader as fdr
```

```python
# 삼성증권
samsung_df = fdr.DataReader('016360', '2020-01-01')
```

```python
stock_date_score.reset_index(inplace=True)
```

```python
stock_date_score.rename(columns = {'date': 'Date'}, inplace = True)
```

```python
stock_date_score.set_index('Date')
```

```python
result = pd.merge(stock_date_score, samsung_df, on='Date', how='right')
```
